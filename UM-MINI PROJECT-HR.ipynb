```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Example 1: Attrition Rate
df=pd.read_csv(r"C:\Users\Mizzy\updated_data.csv")
attrition_counts = df['Attrition'].value_counts()
plt.figure(figsize=(8, 6))
attrition_counts.plot(kind='bar', color='skyblue')
plt.title('Attrition Rate')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Example 2: Age Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Example 3: Attrition Rate by Department
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', hue='Attrition', data=df, palette='coolwarm')
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Attrition')
plt.show()
```

```python
# Example 4: Years of Service Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Years At Company'], bins=15, kde=True, color='orange')
plt.title('Years of Service Distribution')
plt.xlabel('Years at Company')
plt.ylabel('Frequency')
plt.show()

# Example 5: Job Level and Role
plt.figure(figsize=(10, 6))
sns.countplot(x='Job Level', hue='Attrition', data=df, palette='Set2')
plt.title('Attrition Rate by Job Level')
plt.xlabel('Job Level')
plt.ylabel('Count')
plt.legend(title='Attrition', loc='upper right')
plt.show()

# Example 6: Distance from Home
plt.figure(figsize=(8, 6))
sns.histplot(df['Distance From Home'], bins=15, kde=True, color='purple')
plt.title('Distance from Home Distribution')
plt.xlabel('Distance from Home (miles)')
plt.ylabel('Frequency')
plt.show()

# Example 7: Salary and Compensation
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='Monthly Income', data=df, palette='husl')
plt.title('Monthly Income Distribution by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()

# Example 8: Work-Life Balance
plt.figure(figsize=(8, 6))
sns.countplot(x='Work Life Balance', hue='Attrition', data=df, palette='viridis')
plt.title('Attrition Rate by Work-Life Balance')
plt.xlabel('Work-Life Balance')
plt.ylabel('Count')
plt.legend(title='Attrition', loc='upper right')
plt.show()

# Example 9: Performance Rating
plt.figure(figsize=(8, 6))
sns.countplot(x='Performance Rating', hue='Attrition', data=df, palette='muted')
plt.title('Attrition Rate by Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Count')
plt.legend(title='Attrition', loc='upper right')
plt.show()
```

```python

# Conclusion 1: Attrition Rate
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
print("Attrition Rate:")
print(attrition_rate)

# Conclusion 2: Age Distribution
print("\nAge Distribution Summary:")
print(df['Age'].describe())

# Conclusion 3: Job Level
job_level_attrition = df.groupby('Job Level')['Attrition'].value_counts(normalize=True).unstack() * 100
print("\nAttrition Rate by Job Level:")
print(job_level_attrition)

# Conclusion 4: Department
department_attrition = df.groupby('Department')['Attrition'].value_counts(normalize=True).unstack() * 100
print("\nAttrition Rate by Department:")
print(department_attrition)

# Conclusion 5: Work-Life Balance
work_life_balance_attrition = df.groupby('Work Life Balance')['Attrition'].value_counts(normalize=True).unstack() * 100
print("\nAttrition Rate by Work-Life Balance:")
print(work_life_balance_attrition)

# Conclusion 6: Education Level
education_attrition = df.groupby('Education')['Attrition'].value_counts(normalize=True).unstack() * 100
print("\nAttrition Rate by Education Level:")
print(education_attrition)
```
