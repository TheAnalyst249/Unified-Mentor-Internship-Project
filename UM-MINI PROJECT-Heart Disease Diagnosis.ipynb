```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Heart Disease Analysis.csv')
```

```python
print(data.head())
print(data.info())
print(data.describe())
```

```python
# Percentage of patients with chest pain
chest_pain_percentage = (data['Chest Pain Type'] > 0).mean() * 100
print(f"Percentage of patients with chest pain: {chest_pain_percentage:.2f}%")

# Percentage of patients with high blood pressure
high_bp_percentage = (data['Resting Blood Pressure'] > 140).mean() * 100
print(f"Percentage of patients with high blood pressure: {high_bp_percentage:.2f}%")

# Percentage of patients with high cholesterol
high_cholesterol_percentage = (data['Serum Cholesterol Level'] > 200).mean() * 100
print(f"Percentage of patients with high cholesterol: {high_cholesterol_percentage:.2f}%")

# Percentage of patients with diabetes
diabetes_percentage = (data['Fasting Blood Sugar'] > 120).mean() * 100
print(f"Percentage of patients with diabetes: {diabetes_percentage:.2f}%")
```

```python
# Bar plot for chest pain types
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Chest Pain Type')
plt.title('Distribution of Chest Pain Types')
plt.show()
```

```python
# Scatter plot for age vs. cholesterol
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Age', y='Serum Cholesterol Level')
plt.title('Age vs. Cholesterol')
plt.show()
```

```python
# Mean age
mean_age = data['Age'].mean()
print(f"Mean Age: {mean_age:.2f}")

# Median age
median_age = data['Age'].median()
print(f"Median Age: {median_age}")

# Age distribution
age_bins = [20, 30, 40, 50, 60, 70, 80, 90]
age_labels = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89']
data['age_bin'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels, include_lowest=True)
age_dist = data['age_bin'].value_counts(normalize=True) * 100
print(age_dist)
```

```python
#Age Distribution by Chest Pain Type
plt.figure(figsize=(8, 6))
sns.violinplot(x='Chest Pain Type', y='Age', data=data)
plt.title('Age Distribution by Chest Pain Type')
plt.show()
```

```python
# Resting Blood Pressure
plt.figure(figsize=(8, 6))
sns.histplot(data['Resting Blood Pressure'], kde=True)
plt.title('Distribution of Resting Blood Pressure')
plt.show()

#Resting B.P Stats
resting_bp_stats = data['Resting Blood Pressure'].describe()
print('Resting Blood Pressure Statistics:', resting_bp_stats)

plt.figure(figsize=(8, 6))
sns.boxplot(x='Thallium Stress Test', y='Resting Blood Pressure', data=data)
plt.title('Resting Blood Pressure by Heart Disease Status')
plt.show()
```

```python
# Serum Cholesterol
plt.figure(figsize=(8, 6))
sns.histplot(data['Serum Cholesterol Level'], kde=True)
plt.title('Distribution of Serum Cholesterol')
plt.show()

cholesterol_stats = data['Serum Cholesterol Level'].describe()
print('Serum Cholesterol Statistics:', cholesterol_stats)

plt.figure(figsize=(8, 6))
sns.boxplot(x='Thallium Stress Test', y='Serum Cholesterol Level', data=data)
plt.title('Serum Cholesterol by Heart Disease Status')
plt.show()
```

```python
# Fasting Blood Sugar
fasting_blood_sugar_counts = data['Fasting Blood Sugar'].value_counts(normalize=True) * 100
print('Fasting Blood Sugar Distribution:', fasting_blood_sugar_counts)

plt.figure(figsize=(8, 6))
sns.countplot(x='Fasting Blood Sugar', data=data)
plt.title('Distribution of Fasting Blood Sugar')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Fasting Blood Sugar', hue='Thallium Stress Test', data=data)
plt.title('Fasting Blood Sugar by Heart Disease Status')
plt.show()
```

```python
# Resting Electrocardiographic Results
plt.figure(figsize=(8, 6))
sns.countplot(x='Resting ECG', data=data)
plt.title('Distribution of Resting Electrocardiographic Results')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Resting ECG', hue='Thallium Stress Test', data=data)
plt.title('Resting Electrocardiographic Results by Heart Disease Status')
plt.show()
```

```python
# Exercise-induced Angina
exercise_angina_counts = data['Exercise Induced Angina'].value_counts(normalize=True) * 100
print('Exercise-induced Angina Distribution:', exercise_angina_counts)

plt.figure(figsize=(8, 6))
sns.countplot(x='Exercise Induced Angina', data=data)
plt.title('Distribution of Exercise-induced Angina')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Exercise Induced Angina', hue='Thallium Stress Test', data=data)
plt.title('Exercise-induced Angina by Heart Disease Status')
plt.show()
```

```python
# ST Depression Induced by Exercise (oldpeak)
plt.figure(figsize=(8, 6))
sns.histplot(data['ST Depression'], kde=True)
plt.title('Distribution of ST Depression Induced by Exercise')
plt.show()

oldpeak_stats = data['ST Depression'].describe()
print('ST Depression Induced by Exercise Statistics:', oldpeak_stats)

plt.figure(figsize=(8, 6))
sns.boxplot(x='Thallium Stress Test', y='ST Depression', data=data)
plt.title('ST Depression Induced by Exercise by Heart Disease Status')
plt.show()
```

```python
# Slope of the Peak Exercise ST Segment (st_slope)
plt.figure(figsize=(8, 6))
sns.countplot(x='ST Segment Slope', data=data)
plt.title('Distribution of Slope of the Peak Exercise ST Segment')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='ST Segment Slope', hue='Thallium Stress Test', data=data)
plt.title('Slope of the Peak Exercise ST Segment by Heart Disease Status')
plt.show()
```

```python
# Number of Major Vessels (num_vessels)
plt.figure(figsize=(8, 6))
sns.countplot(x='Vessels Colored By Flouoroscopy', data=data)
plt.title('Distribution of Number of Major Vessels')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Vessels Colored By Flouoroscopy', hue='Thallium Stress Test', data=data)
plt.title('Number of Major Vessels by Heart Disease Status')
plt.show()
```

```python
# Maximum Heart Rate
plt.figure(figsize=(8, 6))
sns.histplot(data['Maximum Heart Rate Achieved'], kde=True)
plt.title('Distribution of Maximum Heart Rate')
plt.show()

max_heart_rate_stats = data['Maximum Heart Rate Achieved'].describe()
print('Maximum Heart Rate Statistics:', max_heart_rate_stats)

plt.figure(figsize=(8, 6))
sns.boxplot(x='Thallium Stress Test', y='Maximum Heart Rate Achieved', data=data)
plt.title('Maximum Heart Rate by Heart Disease Status')
plt.show()
```

```python
#Heart Disease Possibility
heart_disease_col = 'Heart Disease Possibility'

# Count the number of 0s and 1s
value_counts = data[heart_disease_col].value_counts()
print('Value Counts:', value_counts)

# Visualize distribution
plt.figure(figsize=(6, 4))
sns.countplot(x=heart_disease_col, data=data)
plt.title(f'Distribution of Heart Disease Possibility')
plt.show()

# Visualize distribution by other attributes
for attr in ['Age', 'Gender', 'Chest Pain Type', 'Resting Blood Pressure', 'Serum Cholesterol Level']:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=heart_disease_col, hue=attr, data=data)
    plt.title(f'Heart Disease Possibility by {attr}')
    plt.show()
```

```python
# Calculate heart disease rates for 0 and 1 values
heart_disease_rates = data.groupby(heart_disease_col).size() / len(data)
print('Heart Disease Rates:', heart_disease_rates)
```
